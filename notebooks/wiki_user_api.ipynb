```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
hamest_file = Path("datas/previous-project/Hedi/hamest.xlsx")
hamest = pd.read_excel(hamest_file)
hamest = hamest.iloc[:, :-2]
hamest = hamest.rename(columns={"Annotation_1": "Hamest"})
hamest = hamest.rename(columns={"Judgment": "ChatGPT"})

hamest_users = []

for record in hamest["JSON_Record"]:
    if pd.isna(record):
        hamest_users.append(None)  # placeholder for missing JSON
    else:
        try:
            data = json.loads(record)
            hamest_users.append(data.get("User", None))
        except Exception as e:
            hamest_users.append(None)  # placeholder for parsing error : 86
```

```python
hamest["User"] = hamest_users
```

```python
user_counts = (
    hamest.groupby("User")["Hamest"]
    .value_counts()
    .unstack(fill_value=0)   # makes columns Weaponised / Not weaponised
)

user_counts.sample(10)
```

```python
user_counts_over1 = user_counts[user_counts.sum(axis=1) > 1]
user_counts_over1
```
