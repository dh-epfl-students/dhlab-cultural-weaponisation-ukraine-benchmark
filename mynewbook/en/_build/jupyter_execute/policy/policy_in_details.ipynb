# Wikipedia Policy in details 

## Human Administration

The Wikimedia Foundation (WMF) is a not-for-profit organization and charity governed by a Board of Trustees (BoT), ~15 people that oversees the foundation and its work, as its ultimate corporate authority. The BoT is not implicated at all in the creation of WP’s policies nor its daily implementation. This task is reserved to the Wikipedia Community (Wikipedians), a self-organizing and self-governing organisation of editors (opposed to the readers that only read WP articles), with various permissions and rights. 

Wikipedians developed the WP policy, a document that can be summarized in five pillars:

- Wikipedia is an encyclopedia  
- Wikipedia is written from a neutral point of view (NPOV)  
- Wikipedia is free content that anyone can use, edit, and distribute  
- Wikipedia’s editors should treat each other with respect and civility  
- Wikipedia has no firm rules


The second pillar “NPOV” is also part of Wikipedia's three core content policies; the other two are "Verifiability" (people can check the facts or claims corresponding to reliable sources) and "No original research" (articles in WP must contain no original research). These policies jointly determine the type and quality of material acceptable in Wikipedia articles, and because they work in harmony, they should not be interpreted in isolation from one another.

To ensure community rules are and policies are respected, a band of trusted users are elected or appointed by the WMF. Among the trusted users, we have the administrators (admins), bureaucrats and functionaries (Stewards and ArbCom).

### Advanced permission groups

- **`autoreviewer`**  
  The user’s edits are automatically marked as *patrolled*, meaning they do not require review by other editors.  
  This permission is typically granted to trusted contributors to reduce the patrolling workload.

- **`reviewer`**  
  Allows the user to review and manage edits under the *Pending Changes* system. Reviewers can:
  - accept or reject pending edits,
  - patrol edits made by other users.  
  This represents a moderate level of editorial responsibility.

- **`rollbacker`**  
  Grants access to the rollback tool, enabling the rapid reversion of consecutive vandalism edits.  
  This permission is primarily used for anti-vandalism activities and requires community trust, but remains below administrator level.

- **`oathauth-twofactorauth`**  
  Indicates that the user has enabled two-factor authentication (2FA) on their account.

## Wikipedia Vital Level 

ithin the English Wikipedia, a subset of articles is designated as [Vital Articles](https://en.wikipedia.org/wiki/Wikipedia:Vital_articles). This initiative identifies the most important topics for the encyclopedia and serves both as a prioritisation framework for editorial improvement and as a centralised watchlist for highly visible content. Vital Articles are organised into five hierarchical levels: Level~1 includes the ten most essential articles, which includes The arts, Human, Philosophy, Life,  Level~2 expands this set to one hundred articles, and each subsequent level progressively broadens the scope by building on the previous one.
